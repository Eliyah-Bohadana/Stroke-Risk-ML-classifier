```python
import pandas as pd
df = pd.read_csv("/Users/eliyahbohadana/Desktop/Stroke-Risk-ML-classifier/data/stroke_raw.csv")
df.shape
```

```python
df.isna().sum().sort_values(ascending=False) 
```

```python
df['bmi']= df['bmi'].fillna(df['bmi'].median())
df.isna().sum().sort_values(ascending=False)
```

```python
#columns that need encoding + different type of objects in different categories
x= df.select_dtypes("object").columns
for i in x : 
   y= df[i].value_counts()
   print(y)
```

#problem: as we look through the different type of answers per category we notice answers #like "unknown" in smoking_status. although "unknown" describes missing information, its #presence is in a lot of rows and is too valuable to delete



```python
df["gender"] = df["gender"].map({"Male": 1, "Female": 0})
df["Residence_type"] = df["Residence_type"].map({"Urban": 1, "Rural": 0})
df["ever_married"] = df["ever_married"].map({"Yes": 1, "No": 0})
```

```python
df["work_type"] = df["work_type"].replace({"Never_worked": "children"})
```

```python
df = pd.get_dummies(df, columns=["smoking_status", "work_type"], drop_first=True)
```
